```python
#2 Desenvolva um algoritmo que auxilie na gestão de estoque de uma loja. O programa deve permitir que o usuário insira o nome e a quantidade de diferentes produtos. Se a quantidade de um produto for inferior a 5 unidades, o sistema deve alertar que o estoque está baixo e sugerir a reposição. O processo deve continuar até que o usuário decida encerrar.
import pandas as pd

estoque = []

print("Sistema de Gestão de Estoque")
print("Digite os produtos e suas quantidades.")
print("Digite 'sair' para encerrar.\n")

while True:
    produto = input("Nome do produto (ou 'sair'): ").strip()

    if produto.lower() == 'sair':
        break

    try:
        quantidade = int(input(f"Quantidade de '{produto}': "))

        if quantidade < 0:
            print("Quantidade não pode ser negativa.")
            continue

        estoque.append({"Produto": produto, "Quantidade": quantidade})

        if quantidade < 5:
            print("Estoque baixo! Reposição recomendada.\n")
        else:
            print("Produto registrado com sucesso.\n")

    except ValueError:
        print("Por favor, digite um número válido.\n")

df = pd.DataFrame(estoque)

print("\n Estoque Atual:")
print(df.to_string(index=False))

df.to_excel("estoque_loja.xlsx", index=False)
print("\n Dados salvos em estoque_loja.xlsx")
```

```python
#3 Elabore um algoritmo que simule um quiz de perguntas e respostas. O programa deve apresentar uma série de perguntas ao usuário, oferecer opções de resposta e verificar se a resposta fornecida está correta. Ao final, o sistema deve exibir a pontuação total do usuário e fornecer um feedback baseado no desempenho.
import pandas as pd

perguntas = [
    {"pergunta": "1. Qual fruta e vermelha por dentro e verde por fora?",
     "opcoes": ["a) Banana", "b) Melancia", "c) Abacate", "d) Abacaxi"],
     "resposta": "b"},
    {"pergunta": "2.Qual é o maior país do mundo em extensão territorial?",
     "opcoes": ["a) China", "b) Estados Unidos", "c) Rússia", "d) Canadá"],
     "resposta": "c"},
    {"pergunta": "3. Qual é o elemento químico representado pelo símbolo O ?",
     "opcoes": ["a) Ouro", "b) Oxigênio", "c) Osmônio", "d) Ozonio"],
     "resposta": "b"},
    {"pergunta": "4. Qual é o planeta mais próximo do Sol?",
     "opcoes": ["a) Terra", "b) Vênus", "c) Mercúrio", "d) Marte"],
     "resposta": "c"}
]

resultados = []
pontuacao = 0

print("Sistema de Quiz - Digite sua resposta (a, b, c ou d):\n")

for questao in perguntas:
    print(questao["pergunta"])
    for opcao in questao["opcoes"]:
        print(opcao)

    resposta = input("Sua resposta: ").lower()
    correta = resposta == questao["resposta"]

    if correta:
        print("Correto!\n")
        pontuacao += 1
    else:
        print(f"Incorreto! A resposta certa era '{questao['resposta']}'\n")

    resultados.append({
        "Pergunta": questao["pergunta"],
        "Sua Resposta": resposta,
        "Correta": questao["resposta"],
        "Acertou?": "Sim" if correta else "Não"})

print(f"Pontuação final: {pontuacao} de {len(perguntas)}")

if pontuacao == len(perguntas):
    print("Excelente!")
elif pontuacao >= len(perguntas) // 2:
    print("Bom desempenho!")
else:
    print("Continue estudando.")

df_resultado = pd.DataFrame(resultados)
df_resultado.to_excel("resultado_quiz.xlsx", index=False)

print("\n Resultados salvos em 'resultado_quiz.xlsx'")
```

```python
#4 Crie um algoritmo que simule um sistema de controle de acesso a um edifício. O programa deve solicitar ao usuário um código de acesso e verificar se ele é válido. Caso contrário, deve permitir novas tentativas até que o número máximo de 3 tentativas seja atingido. Se o acesso for concedido, uma mensagem de boas-vindas deve ser exibida; caso contrário, uma mensagem de bloqueio deve ser apresentada.
import pandas as pd
from datetime import datetime

print("Sistema de Controle de Acesso")
print("Digite o código de acesso. Você tem até 3 tentativas.\n")

codigo_correto = "1234"
tentativas_max = 3
tentativas = 0
registros = []

while tentativas < tentativas_max:
    codigo = input(f"Tentativa {tentativas + 1} - Digite o código: ")
    horario = datetime.now().strftime("%d/%m/%Y %H:%M:%S")

    if codigo == codigo_correto:
        print("Acesso permitido. Bem-vindo ao edifício!\n")
        registros.append({
            "Tentativa": tentativas + 1,
            "Código Digitado": codigo,
            "Resultado": "Acesso Permitido",
            "Data e Hora": horario})
        break
    else:
        print("Código incorreto.")
        tentativas += 1
        registros.append({
            "Tentativa": tentativas,
            "Código Digitado": codigo,
            "Resultado": "Código Incorreto",
            "Data e Hora": horario})

if tentativas == tentativas_max:
    print("\n Acesso bloqueado. Número máximo de tentativas atingido.")
    registros.append({
        "Tentativa": "Final",
        "Código Digitado": "---",
        "Resultado": "Acesso Bloqueado",
        "Data e Hora": datetime.now().strftime("%d/%m/%Y %H:%M:%S")})

df = pd.DataFrame(registros)

print("\n Relatório de Tentativas:")
print(df.to_string(index=False))

df.to_excel("registro_acesso.xlsx", index=False)
print("\nDados salvos em registro_acesso.xlsx")
```

```python
#5 Desenvolva um algoritmo que ajude no planejamento de uma viagem de carro. O programa deve solicitar a distância total a ser percorrida, a capacidade do tanque de combustível do veículo e o consumo médio de combustível (km/l). Com base nessas informações, o sistema deve calcular quantas paradas para reabastecimento serão necessárias, assumindo que o motorista sempre reabastece o tanque completamente.
import pandas as pd
from math import ceil

print("Sistema de Planejamento de Viagem")
print("Digite os dados da viagem para calcular paradas de abastecimento.\n")

try:
    distancia = float(input("Distância total da viagem (km): "))
    capacidade_tanque = float(input("Capacidade do tanque (litros): "))
    consumo_medio = float(input("Consumo médio do veículo (km/l): "))

    autonomia = capacidade_tanque * consumo_medio
    paradas = ceil(distancia / autonomia) - 1
    paradas = max(paradas, 0)

    print("\nCálculo realizado com sucesso.\n")

    if paradas == 0:
        print("Nenhuma parada necessária. Viagem possível com 1 tanque.")
    else:
        print(f"Serão necessárias {paradas} parada(s) para reabastecimento.")

    viagem = [{
        "Distância (km)": distancia,
        "Capacidade do Tanque (L)": capacidade_tanque,
        "Consumo (km/L)": consumo_medio,
        "Autonomia (km)": autonomia,
        "Paradas Necessárias": paradas}]

    df = pd.DataFrame(viagem)

    print("\n Planejamento da Viagem:")
    print(df.to_string(index=False))

    df.to_excel("planejamento_viagem.xlsx", index=False)
    print("\nDados salvos em planejamento_viagem.xlsx")

except ValueError:
    print("Erro: Por favor, digite apenas números válidos.")
```

```python
#6 Elabore um algoritmo que simule um sistema de avaliação de filmes. O programa deve permitir que o usuário insira o título de vários filmes e uma nota de 0 a 10 para cada um. Após o usuário finalizar as entradas, o sistema deve exibir a média das notas atribuídas e listar os filmes que receberam nota superior a 8, destacando-os como altamente recomendados.
import pandas as pd

avaliacoes = []
print("Sistema de Avaliação de Filmes")

while True:
    filme = input("Digite o nome do filme (ou 'sair' para encerrar): ")
    if filme.lower() == 'sair':
        if not avaliacoes:
            print("Nenhum filme foi avaliado!")
        break

    while True:
        try:
            nota = float(input(f"Nota para '{filme}' (0 a 10): "))
            if 0 <= nota <= 10:
                avaliacoes.append({'Filme': filme, 'Nota': nota})
                break
            else:
                print("A nota deve estar entre 0 e 10")
        except:
            print("Digite um valor numérico válido")

df = pd.DataFrame(avaliacoes)
media = df['Nota'].mean()
recomendados = df[df['Nota'] > 8]
print("\nAvaliações dos Filmes:")
print(df.to_string(index=False))

print(f"\nMédia das notas: {media:.1f}")

if not recomendados.empty:
    print("\nFilmes Altamente Recomendados (nota > 8):")
    print(recomendados.to_string(index=False))

df.to_excel("avaliacoes_filmes.xlsx", index=False)
print("\nDados salvos em avaliacoes_filmes.xlsx")
```

```python
#7 Crie um algoritmo que simule um sistema de monitoramento de temperatura ambiente. O programa deve solicitar ao usuário que insira a temperatura registrada a cada hora durante um período de 24 horas. Ao final, o sistema deve calcular a temperatura média do dia, identificar a temperatura mais alta e a mais baixa registradas, e alertar se em algum momento a temperatura ultrapassou 30°C ou ficou abaixo de 15°C.​
import pandas as pd
temperaturas = [
    21, 22, 19, 18, 17, 16, 15, 15, 16, 18, 23, 24,
    27, 30, 32, 33, 32, 30, 28, 26, 24, 25, 22, 21
]

print("RELATÓRIO DE TEMPERATURA - 24 HORAS")
print("")

for hora in range(24):
    print(f"{hora:02d}:00 - {temperaturas[hora]}°C")

media = sum(temperaturas) / len(temperaturas)
maior = max(temperaturas)
menor = min(temperaturas)

print("\nRESUMO DO DIA:")
print(f"Temperatura média: {media:.1f}°C")
print(f"Temperatura mais alta: {maior}°C")
print(f"Temperatura mais baixa: {menor}°C")

if (t > 30 for t in temperaturas):
    print("Alerta: Temperatura ultrapassou 30°C em algum momento.")
else:
    print(" Nenhuma temperatura ultrapassou 30°C.")

if (t < 15 for t in temperaturas):
    print(" Alerta: Temperatura ficou abaixo de 15°C em algum momento.")
else:
    print(" Nenhuma temperatura ficou abaixo de 15°C.")
```

```python
#8 Desenvolva um algoritmo que auxilie na organização de uma biblioteca pessoal. O programa deve permitir que o usuário insira o título e o autor de diversos livros. Após a inserção, o sistema deve oferecer a opção de buscar por livros de um autor específico e listar todos os títulos correspondentes. O processo deve continuar até que o usuário decida encerrar.​
def adicionar_livro(titulo, autor):
    with open('biblioteca.txt', 'a', encoding='utf-8') as arquivo:
        arquivo.write(f"{titulo};{autor}\n")
def buscar_livros_por_autor(autor_busca):
    livros_encontrados = []
    with open('biblioteca.txt', 'r', encoding='utf-8') as arquivo:
        for linha in arquivo:
            titulo, autor = linha.strip().split(';')
            if autor.lower() == autor_busca.lower():
                livros_encontrados.append(titulo)
    return livros_encontrados
def biblioteca_pessoal():
    while True:
        print("\nMenu:")
        print("1. Adicionar livro")
        print("2. Buscar livros por autor")
        print("3. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            titulo = input("Digite o título do livro: ")
            autor = input("Digite o autor do livro: ")
            adicionar_livro(titulo, autor)
            print(f'Livro "{titulo}" de {autor} adicionado com sucesso!')

        elif opcao == '2':
            autor_busca = input("Digite o nome do autor que deseja buscar: ")
            livros = buscar_livros_por_autor(autor_busca)
            if livros:
                print(f'Livros de {autor_busca}:')
                for titulo in livros:
                    print(f'- {titulo}')
            else:
                print(f'Nenhum livro encontrado para o autor {autor_busca}.')

        elif opcao == '3':
            print("Saindo do sistema. Até logo!")
            break

        else:
            print("Opção inválida. Tente novamente.")
biblioteca_pessoal()
```

```python
#9 Elabore um algoritmo que simule um sistema de gerenciamento de tarefas domésticas. O programa deve permitir que o usuário insira diferentes tarefas e atribua um nível de prioridade a cada uma (alto, médio ou baixo). O sistema deve então exibir as tarefas organizadas por ordem de prioridade, começando pelas de alta prioridade, e permitir que o usuário marque tarefas como concluídas.
import pandas as pd
agua_diaria = []
exercicio_diario = []

for dia in range(7):
    agua = int(input(f"Dia {dia + 1} - Copos de água ingeridos: "))
    exercicio = int(input(f"Dia {dia + 1} - Minutos de exercício: "))
    agua_diaria.append(agua)
    exercicio_diario.append(exercicio)

media_agua = sum(agua_diaria) / 7
media_exercicio = sum(exercicio_diario) / 7
7
recomendacao_agua = 8
recomendacao_exercicio = 30

print(f"\nMédia de copos de água por dia: {media_agua:.2f}")
print(f"Média de minutos de exercício por dia: {media_exercicio:.2f}")

if media_agua >= recomendacao_agua:
    print("Você está atendendo à recomendação de água!")
else:
    print("Você deve beber mais água.")

if media_exercicio >= recomendacao_exercicio:
    print("Você está atendendo à recomendação de exercícios!")
else:
    print("Você deve se exercitar mais.")
```

```python
#10 Crie um algoritmo que simule um sistema de acompanhamento de hábitos saudáveis. O programa deve permitir que o usuário registre diariamente a quantidade de copos de água ingeridos e o número de minutos dedicados a atividades físicas. Ao final de uma semana, o sistema deve calcular a média diária de consumo de água e de exercícios, compará-las com as recomendações de saúde (por exemplo, 8 copos de água e 30 minutos de exercício por dia), e fornecer um feedback sobre o desempenho do usuário.
def acompanhar_habitos():
    dias_semana = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]
    agua_total = 0
    exercicio_total = 0

    print("=== Acompanhamento de Hábitos Saudáveis ===")

    for dia in dias_semana:
        print(f"\n--- {dia} ---")
        while True:
            try:
                copos_agua = int(input("Quantos copos de água você bebeu hoje? "))
                if copos_agua < 0:
                    raise ValueError
                break
            except ValueError:
                print("Digite um número válido (maior ou igual a 0).")

        while True:
            try:
                minutos_exercicio = int(input("Quantos minutos de exercício você fez hoje? "))
                if minutos_exercicio < 0:
                    raise ValueError
                break
            except ValueError:
                print("Digite um número válido (maior ou igual a 0).")

        agua_total += copos_agua
        exercicio_total += minutos_exercicio

    media_agua = agua_total / 7
    media_exercicio = exercicio_total / 7

    print("\n=== Relatório Semanal ===")
    print(f"Média de copos de água por dia: {media_agua:.1f}")
    print(f"Média de minutos de exercício por dia: {media_exercicio:.1f}")

    print("\n=== Feedback ===")

    if media_agua >= 8:
        print("Ótimo! Você está se hidratando bem.")
    else:
        print("Atenção! Tente beber pelo menos 8 copos de água por dia.")

    if media_exercicio >= 30:
        print("Parabéns! Você está fazendo exercícios suficientes.")
    else:
        print("Importante! Tente fazer pelo menos 30 minutos de atividade física por dia.")
acompanhar_habitos()
```
